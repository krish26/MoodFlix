# parsing json strings into python objects

```python
def parse_json_column(column):
        """Parse JSON-like string columns"""
        def safe_parse(x):
            try:
                if pd.isna(x):
                    return []
                return literal_eval(x)
            except:
                return []
        return column.apply(safe_parse)
```

```python
movies['genres'] = parse_json_column(movies['genres'])
movies['keywords'] = parse_json_column(movies['keywords'])
movies['cast'] = parse_json_column(movies['cast'])
movies['crew'] = parse_json_column(movies['crew'])
```

Extracting  names from dictonary

```python
def extract_names( items, key='name', limit=None):
        """Extract names from list of dictionaries"""
        if isinstance(items, list):
            names = [item[key] for item in items if key in item]
            return names[:limit] if limit else names
        return []
```

```python
movies['genres']=movies['genres'].apply(lambda x :extract_names(x))
movies['keyword'] = movies['keywords'].apply(lambda x: extract_names(x, limit=10))
movies['cast'] = movies['cast'].apply(lambda x: extract_names(x, limit=5))
```

```python
def get_director(crew):
            if isinstance(crew, list):
                directors = [member['name'] for member in crew if member.get('job') == 'Director']
                return directors[0] if directors else None
            return None
        
movies['director'] = movies['crew'].apply(get_director)
        
```

dropping NA

```python
# Remove movies with no genres
movies = movies[movies['genres'].apply(len) > 0]

# Remove missing overview or rating
movies = movies.dropna(subset=['overview', 'vote_average'])

# Keep quality movies only
movies = movies[movies['vote_count'] >= 50]

print(f"Remaining movies: {len(movies)}")
```

```python
# Filter movies with minimum votes (quality threshold)
movies = movies[movies['vote_count'] >= 50]
```

```python
# Mood to genre mapping
mood_genre_map = {
            'happy': ['Comedy', 'Family', 'Animation', 'Music'],
            'sad': ['Drama', 'Romance'],
            'excited': ['Action', 'Adventure', 'Science Fiction'],
            'scared': ['Horror', 'Thriller'],
            'romantic': ['Romance', 'Drama'],
            'thoughtful': ['Documentary', 'Drama', 'History'],
            'adventurous': ['Adventure', 'Action', 'Fantasy'],
            'relaxed': ['Comedy', 'Family', 'Animation'],
            'mysterious': ['Mystery', 'Thriller', 'Crime'],
            'inspired': ['Documentary', 'Biography', 'History'],
        }
# Mood to keyword hints (these help refine recommendations)
mood_keyword_hints = {
            'happy': ['comedy', 'friendship', 'love', 'celebration', 'fun'],
            'sad': ['loss', 'tragedy', 'melancholy', 'death', 'breakup'],
            'excited': ['explosion', 'chase', 'hero', 'battle', 'victory'],
            'scared': ['ghost', 'monster', 'haunted', 'murder', 'supernatural'],
            'romantic': ['love', 'relationship', 'wedding', 'romance', 'couple'],
            'thoughtful': ['philosophy', 'life', 'society', 'truth', 'meaning'],
            'adventurous': ['journey', 'exploration', 'quest', 'treasure', 'travel'],
            'relaxed': ['slice of life', 'peaceful', 'calm', 'nature', 'family'],
            'mysterious': ['detective', 'investigation', 'conspiracy', 'secret', 'puzzle'],
            'inspired': ['success', 'achievement', 'courage', 'overcoming', 'motivation'],
        }
```

```python
print("Calculating mood scores...")

for mood in mood_genre_map.keys():
    scores = []
    preferred_genres = mood_genre_map[mood]
    keyword_hints = mood_keyword_hints.get(mood, [])
    
    for _, row in movies.iterrows():
        score = 0
        
        # Genre matching
        genre_matches = len(set(row['genres']) & set(preferred_genres))
        score += genre_matches * 3
        
        # Keyword matching
        movie_keywords = [k.lower() for k in row['keyword']]
        keyword_matches = sum(
            1 for hint in keyword_hints 
            if any(hint in kw for kw in movie_keywords)
        )
        score += keyword_matches * 1.5
        
        # Rating boost
        score += (row['vote_average'] / 10) * 2
        
        # Popularity boost
        if row['popularity'] > 10:
            score += 0.5
        
        scores.append(score)
    
    movies[f'mood_{mood}_score'] = scores

print("Mood scoring complete.")
```

```python
columns_to_keep = [
    'id', 'title', 'overview', 'genres',
    'vote_average', 'vote_count', 'popularity',
    'release_date', 'runtime',
    'cast_names', 'director', 'poster_path'
]

mood_columns = [col for col in movies.columns if col.startswith('mood_')]
columns_to_keep.extend(mood_columns)

columns_to_keep = [col for col in columns_to_keep if col in movies.columns]

final_df = movies[columns_to_keep].copy()

output_path = 'processed_movies.pkl'

with open(output_path, 'wb') as f:
    pickle.dump(final_df, f)

print(f"Saved {len(final_df)} movies")
```
